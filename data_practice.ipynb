```python
import pandas as pd

# Let's check if pandas works perfectly in our new file!
sample_data = {
    'Product': ['Bikes', 'Accessories', 'Clothing', 'Bikes'],
    'Revenue': [1200, 450, 300, 1500],
    'Country': ['Ethiopia', 'Canada', 'Ethiopia', 'Canada']
}

df = pd.DataFrame(sample_data)
df
```

```python
import matplotlib.pyplot as plt

# Group the data by Product and sum the Revenue
product_revenue = df.groupby('Product')['Revenue'].sum()

# Create a bar chart
plt.figure(figsize=(6, 4))
product_revenue.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])

# Add labels and a title
plt.title('Total Revenue by Product Category', fontsize=14)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.xticks(rotation=0)

# Render the graph
plt.show()
```
